# Local Setup (VSCode / Anaconda version)

This replaces `colab_setup.ipynb` for anyone **not** using Google Colab.

### What's different from the Colab version
- No `from google.colab import drive` — that only exists inside Colab, it will
  always crash on a local machine. We use a normal local folder instead.
- No Google Drive mounting. You just put the CSVs in a folder next to your notebooks.
- Same end result: `TRAIN_PATH` and `TEST_PATH` pointing at usable CSVs.

### What you need to do before running this
1. Download the data from the [Kaggle competition page](https://www.kaggle.com/competitions/jigsaw-toxic-comment-classification-challenge/data)
   (you'll need a free Kaggle account, click "Download All").
2. Unzip it. You should end up with `train.csv`, `test.csv`, `test_labels.csv`,
   `sample_submission.csv` (Kaggle sometimes nests these in their own
   sub-zips — if so, unzip those too).
3. In your project folder, create a `data/` folder and put all four CSVs directly
   inside it (not in subfolders). Your project should look like:

```
toxic-comment-classification/
├── data/
│   ├── train.csv
│   ├── test.csv
│   ├── test_labels.csv
│   └── sample_submission.csv
├── notebooks/
│   ├── 00_local_setup.ipynb   <- this file
│   ├── 01_data_exploration.ipynb
│   ├── 02_eda.ipynb
│   └── 03_tokenization.ipynb
├── models/
└── ...
```

If your `data/` folder ends up nested differently, just edit `DATA_DIR` below
to match wherever your CSVs actually are.


## Step 1 — Set the data folder path

````python
import os
import pandas as pd

# This assumes your notebook lives in a "notebooks/" folder, one level
# below the project root, and your data lives in "data/" at the project root.
# Adjust if your folder layout is different.
DATA_DIR = os.path.join("..", "data")

TRAIN_PATH = os.path.join(DATA_DIR, "train.csv")
TEST_PATH = os.path.join(DATA_DIR, "test.csv")
TEST_LABELS_PATH = os.path.join(DATA_DIR, "test_labels.csv")
````

## Step 2 — Sanity check the files actually exist

````python
for name, path in [("train.csv", TRAIN_PATH), ("test.csv", TEST_PATH), ("test_labels.csv", TEST_LABELS_PATH)]:
    if os.path.exists(path):
        print(f"FOUND     {name:18s} -> {path}")
    else:
        print(f"MISSING   {name:18s} -> {path}  (check DATA_DIR above)")
````

## Step 3 — Load and verify shapes

````python
train = pd.read_csv(TRAIN_PATH)
test = pd.read_csv(TEST_PATH)

print(f"train shape: {train.shape}")
print(f"test shape : {test.shape}")
print("Dataset is ready.")
````
